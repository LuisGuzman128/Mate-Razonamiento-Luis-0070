#### problema 1
La suma de un número más 9 es igual a 21. ¿Cuál es ese número?Paso 1: Análisis y ResoluciónDefinir la variable: Sea $x$ el número que buscamos.Plantear la ecuación: El enunciado dice "un número ($x$) más ($+$) el número $9$ es igual ($=$) a $21$".$$x + 9 = 21$$Despejar la variable: Para dejar a la $x$ sola, restamos 9 en ambos lados de la igualdad:$$x = 21 - 9$$Resultado:$$x = 12$$

```python
# Problema 1: Encontrar el número desconocido
# resultado_total = 21
# mero_sumado = 9
resultado_total=int(input("Ingrese el resultado total: "))
numero_sumado=int(input("Ingrese el número sumado: "))

# Despejamos: x = 21 - 9
x = resultado_total - numero_sumado

print(f"El número buscado en el problema 1 es: {x}")
```

#### problema 2
Definir las variables:Edad de Ana = $x$Edad de Pedro = $2x$ (porque es el doble que Ana)Plantear la ecuación: La suma de ambas edades es 36.$$x + 2x = 36$$Simplificar: Sumamos las términos semejantes ($1x + 2x = 3x$).$$3x = 36$$Despejar la variable: El 3 que multiplica a $x$ pasa al otro lado dividiendo:$$x = \frac{36}{3}$$$$x = 12$$Resultado:Ana ($x$): 12 años.Pedro ($2x$): $2 \times 12 = 24$ años.

```python
# Problema 2: Edades de Pedro y Ana
# suma_total = 36
suma_total=int(input("Ingrese la suma total: "))

# La ecuación es x + 2x = 36 -> 3x = 36
# Dividimos el total entre 3 para hallar la unidad (edad de Ana)
edad_ana = suma_total / 3
edad_pedro = edad_ana * 2

print(f"Ana tiene {int(edad_ana)} años.")
print(f"Pedro tiene {int(edad_pedro)} años.")
```
