Ejercicios Ejercicio 1: $x + 8 = 15$Procedimiento:$$x = 15 - 8$$$$x = 7$$

```python
# Problema 1: x + 8 = 15
# x = 15 - 8
resultado_suma = int(input("Ingrese el resultado de la suma (15): "))
valor_sumado = int(input("Ingrese el valor que suma (8): "))

# Despeje: x = 15 - 8
x = resultado_suma - valor_sumado

print(f"El valor de x es: {x}")
```

ejercicio 3: $3x + 4x = 21$Procedimiento:$$7x = 21$$$$x = 21 / 7$$$$x = 3$$

```python
# Problema 3: 3x + 4x = 21
# 7x = 21
resultado_total = int(input("Ingrese el resultado total (21): "))
suma_coeficientes = int(input("Ingrese la suma de los coeficientes de x (7): "))

# Despeje: x = 21 / 7
x = resultado_total / suma_coeficientes

print(f"El valor de x es: {int(x)}")
```

Ejercicio 5: $6a - 2a + a = 20$Procedimiento:$$4a + a = 20$$$$5a = 20$$$$a = 20 / 5$$$$a = 4$$

```python
# Problema 5: 6a - 2a + a = 20
# 5a = 20
resultado_total = int(input("Ingrese el resultado total (20): "))
coeficiente_final = int(input("Ingrese el coeficiente de 'a' simplificado (5): "))

# Despeje: a = 20 / 5
a = resultado_total / coeficiente_final

print(f"El valor de a es: {int(a)}")
```

Ejercicio 7: $4x + 3 = 2x + 11$Procedimiento:$$4x - 2x = 11 - 3$$$$2x = 8$$$$x = 8 / 2$$$$x = 4$$

```python
# Problema 7: 4x + 3 = 2x + 11
# 4x - 2x = 11 - 3 -> 2x = 8
resta_numeros = int(input("Ingrese la resta de los números (11 - 3 = 8): "))
resta_coeficientes = int(input("Ingrese la resta de las x (4 - 2 = 2): "))

# Despeje: x = 8 / 2
x = resta_numeros / resta_coeficientes

print(f"El valor de x es: {int(x)}")
```

Ejercicio 9: $9m + 4 = 5m + 20$Procedimiento:$$9m - 5m = 20 - 4$$$$4m = 16$$$$m = 16 / 4$$$$m = 4$$

```python
# Problema 9: 9m + 4 = 5m + 20
# 9m - 5m = 20 - 4 -> 4m = 16
resta_numeros = int(input("Ingrese la resta de los números (20 - 4 = 16): "))
resta_coeficientes = int(input("Ingrese la resta de las m (9 - 5 = 4): "))

# Despeje: m = 16 / 4
m = resta_numeros / resta_coeficientes

print(f"El valor de m es: {int(m)}")
```

Ejercicio 11: $6a + 5 = 2a + 21$Procedimiento:$$6a - 2a = 21 - 5$$$$4a = 16$$$$a = 16 / 4$$$$a = 4$$

```python
# Problema 11: 6a + 5 = 2a + 21
# 6a - 2a = 21 - 5 -> 4a = 16
resta_numeros = int(input("Ingrese la resta de los números (21 - 5 = 16): "))
resta_coeficientes = int(input("Ingrese la resta de las a (6 - 2 = 4): "))

# Despeje: a = 16 / 4
a = resta_numeros / resta_coeficientes

print(f"El valor de a es: {int(a)}")
```
