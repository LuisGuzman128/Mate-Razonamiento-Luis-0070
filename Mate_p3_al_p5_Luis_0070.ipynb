#### problema 4

```python
# Problema 4: El triple de un número más 4 es 31
# total = 31
# aumento = 4
# multiplo = 3
total = int(input("Ingrese el total: "))
aumento = int(input("Ingrese el aumento: "))

# Despeje: x = (31 - 4) / 3
x = (total - aumento) / multiplo

print(f"El número del problema 4 es: {int(x)}")
```

#### problema 5
Simplificamos sumando las $x$ y los números:$$4x + 10 = 50$$Despejamos: Pasamos el 10 restando:$$4x = 50 - 10$$$$4x = 40$$Dividimos:$$x = \frac{40}{4}$$$$x = 10$$

```python
# Problema 5: Un terreno tiene un largo que mide 5 metros más que el ancho. Perímetro total 50.
perimetro_total = int(input("Ingrese el perímetro total (ej. 50): "))
diferencia_largo = int(input("Ingrese cuántos metros más mide el largo que el ancho (ej. 5): "))

# El "exceso" de metros en la suma total es 2 veces la diferencia del largo
# Despeje: 4x = perimetro - (diferencia * 2)
ancho = (perimetro_total - (diferencia_largo * 2)) / 4
largo = ancho + diferencia_largo

print(f"El ancho del terreno es: {ancho} metros")
print(f"El largo del terreno es: {largo} metros")
```
